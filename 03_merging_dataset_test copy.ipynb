```python
import pandas as pd

df1 = pd.read_csv('./file_created_2/dki1.csv')
df2 = pd.read_csv('./file_created_2/dki2.csv')
df3 = pd.read_csv('./file_created_2/dki3.csv')
df4 = pd.read_csv('./file_created_2/dki4.csv')
df5 = pd.read_csv('./file_created_2/dki5.csv')
```

```python
df_merged = pd.concat([df1, df2, df3, df4, df5], ignore_index=True)
df_merged.tail()
```

```python
sorted_merged = df_merged.sort_values(by=['tanggal', 'stasiun'], ascending=[True, True])
```
